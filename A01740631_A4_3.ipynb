# 1. Compute Statistics

**Programa Funcional**

```python
"""
Compute Statistics.
"""
import time
from collections import Counter
from google.colab import files


def compute_statistics(numbers):
    """Calcula estadísticas básicas a partir de una lista de números."""
    numbers_sorted = sorted(numbers)
    n = len(numbers)

    mean = sum(numbers) / n if n > 0 else 0
    median = (numbers_sorted[n // 2] if n % 2 != 0 else
              sum(numbers_sorted[n // 2 - 1:n // 2 + 1]) / 2) if n > 0 else 0
    mode = [number for number, freq in Counter(numbers).items()
            if freq == max(Counter(numbers).values())]

    variance = sum((x - mean) ** 2 for x in numbers) / n if n > 0 else 0
    std_dev = variance ** 0.5

    return mean, median, mode, std_dev, variance


def read_numbers_from_file(uploaded_file):
    """Lee los números y los devuelve como una lista de enteros."""
    numbers = []
    for line in uploaded_file.decode('utf-8').splitlines():
        try:
            numbers.append(float(line.strip()))
        except ValueError:
            pass  # Ignora líneas que no se pueden convertir a float
    return numbers


def main():
    """Carga, cálculo y visualización de resultados."""
    start_time = time.time()

    print("Por favor, suba el archivo:")
    uploaded = files.upload()  # Solo una llamada a files.upload() aquí
    file_name = next(iter(uploaded))
    numbers = read_numbers_from_file(uploaded[file_name])
    statistics = compute_statistics(numbers)

    result_file_name = "StatisticsResults.txt"
    with open(result_file_name, 'w', encoding='utf-8') as file:
        file.write(f"Mean: {statistics[0]}\n")
        file.write(f"Median: {statistics[1]}\n")
        mode = ', '.join(map(str, statistics[2])) if statistics[2] else 'None'
        file.write(f"Mode: {mode}\n")
        file.write(f"Standard Deviation: {statistics[3]}\n")
        file.write(f"Variance: {statistics[4]}\n")

    end_time = time.time()
    print(f"Tiempo transcurrido: {end_time - start_time:.2f} segundos. "
          f"Resultados guardados en {result_file_name}")

    # Imprimir el contenido del archivo de resultados
    with open(result_file_name, 'r', encoding='utf-8') as file:
        content = file.read()
        print("Contenido del archivo de resultados:")
        print(content)

    # Descargar el archivo de resultados
    files.download(result_file_name)


if __name__ == "__main__":
    main()
```

# 2. Converter

**Programa Funcional**

```python
"""
Converter: de enteros a sus representaciones binarias y hexadecimales.
"""
import time
from google.colab import files


def convert_to_binary(number):
    """Convierte un número entero a su representación binaria."""
    return bin(number)[2:]


def convert_to_hexadecimal(number):
    """Convierte un número entero a su representación hexadecimal."""
    return hex(number)[2:].upper()


def read_numbers_from_file(uploaded_file):
    """Lee y los devuelve como una lista de enteros."""
    numbers = []
    for line in uploaded_file.decode('utf-8').splitlines():
        numbers.append(int(line.strip()))
    return numbers


def convert_numbers(numbers):
    """De enteros a representaciones binarias y hexadecimales."""
    results = []
    for number in numbers:
        binary = convert_to_binary(number)
        hexadecimal = convert_to_hexadecimal(number)
        results.append((number, binary, hexadecimal))
    return results


def main():
    """Carga de archivos, conversión y visualización de resultados."""
    start_time = time.time()

    uploaded = files.upload()
    file_name = next(iter(uploaded))
    numbers = read_numbers_from_file(uploaded[file_name])
    results = convert_numbers(numbers)

    result_file_name = "ConversionResults.txt"
    with open(result_file_name, 'w', encoding='utf-8') as file:
        for number, binary, hexadecimal in results:
            file.write(f"Number: {number}\n"
                       f"Binary: {binary}\n"
                       f"Hexadecimal: {hexadecimal}\n"
                       "-----\n")

    end_time = time.time()
    print(f"Time elapsed: {end_time - start_time:.2f} seconds. "
          f"Results saved in {result_file_name}")

    # Imprimir el contenido del archivo de resultados
    with open(result_file_name, 'r', encoding='utf-8') as file:
        content = file.read()
        print("Contenido del archivo de resultados:")
        print(content)

    # Descargar el archivo de resultados
    files.download(result_file_name)


if __name__ == "__main__":
    main()
```

# 3. Count Words

**Programa Funcional**

```python
"""
Contador de palabras.
"""
import time
from collections import Counter
from google.colab import files


def read_words(file_path):
    """
    Lee un archivo de texto y devuelve una lista de palabras en minúsculas.
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read().lower().split()


def count_words(words):
    """
    Cuenta la frecuencia de cada palabra en la lista de palabras.
    """
    return Counter(words)


def main_function():
    """
    Carga el archivo, cuenta las palabras y guarda los resultados.
    """
    start_time = time.time()

    # Cargar el archivo desde el sistema local
    uploaded = files.upload()
    file_name = next(iter(uploaded))
    words = read_words(file_name)

    word_counts = count_words(words)

    # Calcular el total de palabras
    total_words = sum(word_counts.values())

    result_file_name = "WordCountResults.txt"
    with open(result_file_name, 'w', encoding='utf-8') as file:
        for word, count in word_counts.items():
            file.write(f"{word}: {count}\n")
        # Agregar el total de palabras al final del archivo
        file.write(f"\nTotal de palabras: {total_words}")

    end_time = time.time()
    print(f"Completado en {end_time - start_time:.2f} segundos. "
          f"Resultados guardados en {result_file_name}")

    # Imprimir el contenido del archivo
    with open(result_file_name, 'r', encoding='utf-8') as file:
        content = file.read()
        print("Contenido del archivo de resultados:")
        print(content)

    # Descargar el archivo de resultados
    files.download(result_file_name)


if __name__ == "__main__":
    main_function()
```
